##Simulate the process on checking the consistency betweent two datasets

```python
import pandas as pd
import numpy as np
import re
import seaborn
seaborn.set()
import matplotlib.pyplot as plt
%matplotlib inline 
import random
from random import randint
```

```python
#Initialize the orginal data set by random
x = np.random.random_integers(1000, size=(100,1))
y = []
standard = x

#seet y to be close to x with a distributed noise
for i in x[:]:
    if randint(0,4) > 3:
        y.append( i - randint(0,100))
    else:                            # y has a high probablity to be higher than x 
        y.append( i + randint(0,100))
     

data_A = pd.DataFrame(x, columns=list(['Column']))
data_B = pd.DataFrame(y, columns=list(['Column']))
plt.scatter(data_A, data_B, color = 'coral', marker = 'D' )
plt.plot(x,standard, color = 'gray')
plt.xlabel('Data_A')
plt.ylabel('Data_B')
plt.rcParams['figure.figsize']=[25,10]
plt.show()
```
